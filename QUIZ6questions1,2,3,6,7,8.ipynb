```python
#Q7
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mm = 0.001
k = 1e7  # wave number in m^-1, adjust for mm scale

# Point sources (positions in mm, scaled)
r1 = np.array([-1, 2, 0]) * mm
e1 = 0.85
r2 = np.array([-1, 10, 1]) * mm
e2 = 0.7
r3 = np.array([-1, 8, -1]) * mm
e3 = 0.7
r4 = np.array([-1, -0.5]) * mm
e4 = 1.0
r5 = np.array([-1, 0.5]) * mm
e5 = 1.0
r6 = np.array([-1, 2, 0]) * mm
e6 = 0.85

# Reference wave
rref = np.array([-10, 0]) * mm
eref = 1.0

# Screen points (rs = (0, ys)) from question 6
n_points = 8000
ys = -40 * mm + (80 * mm / n_points) * np.arange(n_points)  # ys[i] = -40 mm + (80 mm / 8000) * i
interferogram = np.zeros(n_points)

# Function to calculate distance
def distance(r1, r2):
    r1 = np.pad(r1, (0, max(0, len(r2) - len(r1))), mode='constant')
    r2 = np.pad(r2, (0, max(0, len(r1) - len(r2))), mode='constant')
    return np.sqrt(np.sum((r1 - r2)**2))

# Step 1: Recalculate the interferogram (same as question 6)
for i in range(n_points):
    rs = np.array([0, ys[i]])
    E_total = (e1 * np.exp(-1j * k * distance(rs, r1)) +
               e2 * np.exp(-1j * k * distance(rs, r2)) +
               e3 * np.exp(-1j * k * distance(rs, r3)) +
               e4 * np.exp(-1j * k * distance(rs, r4)) +
               e5 * np.exp(-1j * k * distance(rs, r5)) +
               e6 * np.exp(-1j * k * distance(rs, r6)) +
               eref * np.exp(-1j * k * distance(rs, rref)))
    interferogram[i] = np.abs(E_total)**2

# Step 2: Illuminate with the reference beam
U = np.zeros(n_points, dtype=complex)
for i in range(n_points):
    rs = np.array([0, ys[i]])
    E_ref = eref * np.exp(-1j * k * distance(rs, rref))
    U[i] = interferogram[i] * E_ref

# Step 3: Reconstruction points (rf = (1, yf))
n_yf = 31
yf = np.linspace(-1.5 * mm, 1.5 * mm, n_yf)  # yf from -1.5 mm to 1.5 mm
hologram_intensity = np.zeros(n_yf)

# Step size for integration
delta_ys = (80 * mm) / n_points  # 0.01 mm

# Step 4: Propagate the field to rf
for j in range(n_yf):
    rf = np.array([1 * mm, yf[j]])
    E_rf = 0
    for i in range(n_points):
        rs = np.array([0, ys[i]])
        r = distance(rf, rs)
        # Simplified Rayleigh-Sommerfeld propagation (1D approximation)
        E_rf += U[i] * np.exp(-1j * k * r) / r * delta_ys
    hologram_intensity[j] = np.abs(E_rf)**2

# Step 5: Plot the hologram intensity (equivalent to Plot command)
plt.plot(yf / mm, hologram_intensity)  # Convert yf to mm for plotting
plt.xlabel('yf (mm)')
plt.ylabel('|E(yf)|^2')
plt.title('Hologram Intensity at rf = (1, yf)')
plt.grid(True)
plt.show()
```

```python
#Q8
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
mm = 0.001
k = 1e7  # wave number in m^-1, adjust for mm scale

# Point sources (positions in mm, scaled)
r1 = np.array([-1, 2, 0]) * mm
e1 = 0.85
r2 = np.array([-1, 10, 1]) * mm
e2 = 0.7
r3 = np.array([-1, 8, -1]) * mm
e3 = 0.7
r4 = np.array([-1, -0.5]) * mm
e4 = 1.0
r5 = np.array([-1, 0.5]) * mm
e5 = 1.0
r6 = np.array([-1, 2, 0]) * mm
e6 = 0.85

# Reference wave
rref = np.array([-10, 0]) * mm
eref = 1.0

# Step 1: Recalculate the interferogram (from question 6)
n_points = 8000
ys = -40 * mm + (80 * mm / n_points) * np.arange(n_points)  # ys[i] = -40 mm + (80 mm / 8000) * i
interferogram = np.zeros(n_points)

def distance(r1, r2):
    r1 = np.pad(r1, (0, max(0, len(r2) - len(r1))), mode='constant')
    r2 = np.pad(r2, (0, max(0, len(r1) - len(r2))), mode='constant')
    return np.sqrt(np.sum((r1 - r2)**2))

for i in range(n_points):
    rs = np.array([0, ys[i]])
    E_total = (e1 * np.exp(-1j * k * distance(rs, r1)) +
               e2 * np.exp(-1j * k * distance(rs, r2)) +
               e3 * np.exp(-1j * k * distance(rs, r3)) +
               e4 * np.exp(-1j * k * distance(rs, r4)) +
               e5 * np.exp(-1j * k * distance(rs, r5)) +
               e6 * np.exp(-1j * k * distance(rs, r6)) +
               eref * np.exp(-1j * k * distance(rs, rref)))
    interferogram[i] = np.abs(E_total)**2

# Step 2: Illuminate with the reference beam
U = np.zeros(n_points, dtype=complex)
for i in range(n_points):
    rs = np.array([0, ys[i]])
    E_ref = eref * np.exp(-1j * k * distance(rs, rref))
    U[i] = interferogram[i] * E_ref

# Step 3: Reconstruction grid (rf = (xf, yf))
xf = np.arange(-16, 4 + 2, 2) * mm  # -16 mm to 4 mm, step 2 mm
yf = np.arange(-1.5, 1.5 + 0.5, 0.5) * mm  # -1.5 mm to 1.5 mm, step 0.5 mm
XF, YF = np.meshgrid(xf, yf)
hologram_intensity = np.zeros((len(yf), len(xf)))

# Step size for integration
delta_ys = (80 * mm) / n_points  # 0.01 mm

# Step 4: Propagate the field to rf
for j in range(len(yf)):
    for k in range(len(xf)):
        rf = np.array([xf[k], yf[j]])
        E_rf = 0
        for i in range(n_points):
            rs = np.array([0, ys[i]])
            r = distance(rf, rs)
            E_rf += U[i] * np.exp(-1j * k * r) / r * delta_ys
        hologram_intensity[j, k] = np.abs(E_rf)**2

# Step 5: Plot the hologram intensity (equivalent to DiscretePlot3D)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
XF_mm, YF_mm = XF / mm, YF / mm  # Convert to mm for plotting
surf = ax.plot_surface(XF_mm, YF_mm, hologram_intensity, cmap='viridis')  # DarkRainbow equivalent
ax.set_xlabel('xf (mm)')
ax.set_ylabel('yf (mm)')
ax.set_zlabel('Intensity |E|^2')
fig.colorbar(surf, label='Intensity')
plt.title('2D Hologram Intensity at rf = (xf, yf)')
plt.show()
```
